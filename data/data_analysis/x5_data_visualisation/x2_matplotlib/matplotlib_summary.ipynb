# Matplotlib: Summary



```python
import numpy as np
import matplotlib.pyplot as plt
```

## Basic Line Plot

```python
simple_title = "Simple Line Plot"
x_label = "x-axis"
y_label = "y-axis"

x_data = np.array( [ 1, 8 ] )
y_data = np.array( [ 3, 10 ] )

plt.plot( x_data, y_data )

plt.grid( color = "gray", alpha = 0.5 )
plt.tight_layout( )
plt.title( f"{simple_title}" )
plt.xlabel( f"{x_label}" )
plt.ylabel( f"{y_label}" )

plt.show( )
```

## Save Plots 

```python
import numpy as np 
import matplotlib.pyplot as plt

simple_title = "Simple Line Plot"
x_label = "x-axis"
y_label = "y-axis"

x_data = np.array( [ 1, 8 ] )
y_data = np.array( [ 3, 10 ] )

plt.plot( x_data, y_data )

plt.grid( color = "gray", alpha = 0.5 )
plt.tight_layout( )
plt.title( f"{simple_title}" )
plt.xlabel( f"{x_label}" )
plt.ylabel( f"{y_label}" )

plt.tight_layout()
plt.savefig(
  "test_image.jpg",
  bbox_inches='tight',
  pad_inches=0,
)

plt.show( )
```

## Custom Markers

```python
x_data = [1, 2, 3]
y_data = [4, 5, 1]


fig, ax = plt.subplots()
plt.plot(
  x_data,
  y_data,
  marker='o',
  linestyle='--'
)

plt.xticks( np.arange( np.min(x_data), np.max(x_data)+1, 0.5 ) )
plt.yticks( np.arange( np.min(y_data), np.max(y_data)+1, 1 ) )

plt.title( f"{simple_title}" )
plt.xlabel( f"{x_label}" )
plt.ylabel( f"{y_label}" )
plt.grid( color = "gray", alpha = 0.5 )
plt.tight_layout( )


plt.show()
```

## Errors Bars

```python
# example data
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)

err = 0.4

fig, ax = plt.subplots()
ax.errorbar(x, y, yerr=err)
ax.fill_between(x, y+err, y-err, alpha=0.25 )

plt.show()
```

## Replot

```python
x = np.linspace(0, 10, 100)

for i in range(1, 5):
  y = np.sin(x * i)  # Example: generate a new series for each iteration
  plt.plot(x, y, label=f"Series {i}")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Multiple Series in a Single Plot')
# 'upper left', 'upper right', 'lower left', 'lower right'
plt.legend( loc='upper right' )
plt.show()
```

## Heatmaps

```python
def example_plot(ax, fontsize=12, hide_labels=False):
  
  pc = ax.pcolormesh(np.random.randn(30, 30), vmin=-2.5, vmax=2.5)

  if not hide_labels:
    ax.set_xlabel('x-label', fontsize=fontsize)
    ax.set_ylabel('y-label', fontsize=fontsize)
    ax.set_title('Title', fontsize=fontsize)
  return pc

np.random.seed( 19680808 )

fig = plt.figure( layout='constrained', figsize=(16, 6) )
subfigs = fig.subfigures(1, 2, wspace=0.07)

axsLeft = subfigs[0].subplots(1, 2, sharey=True)
# subfigs[0].set_facecolor('0.75')
for ax in axsLeft:
    pc = example_plot(ax)
subfigs[0].suptitle('Left plots', fontsize='x-large')
subfigs[0].colorbar( pc, shrink=0.6, ax=axsLeft, location='bottom' )

axsRight = subfigs[1].subplots(3, 1, sharex=True)
for nn, ax in enumerate(axsRight):
  pc = example_plot(ax, hide_labels=True)
  if nn == 2:
    ax.set_xlabel('xlabel')
  if nn == 1:
    ax.set_ylabel('ylabel')

# subfigs[1].set_facecolor('0.85')
subfigs[1].colorbar( pc, shrink=0.6, ax=axsRight )
subfigs[1].suptitle('Right plots', fontsize='x-large')

fig.suptitle('Figure suptitle', fontsize='xx-large')

plt.show()
```
